```python
import pandas as pd

data = {
    'Candidate': [
        'Ravi',
        'Anjali',
        'Kiran',
        'Rahul',
        'Priya'
    ],
    'Resume': [
        'Python SQL Machine Learning Data Analysis',
        'Java HTML CSS JavaScript Web Development',
        'Python Machine Learning Deep Learning AI',
        'SQL Database Management Excel Power BI',
        'Python SQL Data Analysis Power BI'
    ]
}

df = pd.DataFrame(data)

df
```

```python
job_description = "Python SQL Machine Learning Data Analysis"

print(job_description)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [job_description] + df['Resume'].tolist()

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

similarity_scores = cosine_similarity(
    tfidf_matrix[0:1],
    tfidf_matrix[1:]
)

df['Match Score'] = similarity_scores[0] * 100

df
```

```python
ranked_candidates = df.sort_values(
    by='Match Score',
    ascending=False
)

ranked_candidates
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    ranked_candidates['Candidate'],
    ranked_candidates['Match Score']
)

plt.title("Candidate Match Scores")
plt.xlabel("Candidate")
plt.ylabel("Score")

plt.show()
```
